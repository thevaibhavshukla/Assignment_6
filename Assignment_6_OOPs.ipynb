```python
'''Q1                     '''





# Python program to create Bankaccount class

# with both a deposit() and a withdraw() function

class Bank_Account:

    def __init__(self,account_number,holders_name,balance):

        self.account_number = account_number

        self.holders_name = holders_name

        self.balance=balance

        print("Hello!!! Welcome to the Deposit & Withdrawal Machine")



    def deposit(self):

        amount=float(input("Enter amount to be Deposited: "))

        self.balance += amount

        print("\n Amount Deposited:",amount)



    def withdraw(self):

        amount = float(input("Enter amount to be Withdrawn: "))

        if self.balance>=amount:

            self.balance-=amount

            print("\n You Withdrew:", amount)

        else:

            print("\n Insufficient balance ")



    def display(self):

        print("\n Net Available Balance=",self.balance)



# Driver code



# creating an object of class

wolverine = Bank_Account(77,"wolverine",899999)



# Calling functions with that class object

wolverine.deposit()

wolverine.withdraw()

wolverine.display()

 
```

```python
 

'''Q2                 '''



class Employee:

    def __init__(self, name, number, department, job,salary):

        self.name = name

        self.number = number

        self.department = department

        self.job = job

        self.salary = salary



    def set_name(self, name):

        self.name = name

    def set_number(self, number):

        self.number = number

    def set_department (self, department):

        self.department = department

    def set_job (self, job):

        self.job = job

    def set_salary(self,salary):

        self.salary = salary

    def get_name(self):

        return self.name

    def get_number(self):

        return self.number

    def get_department(self):

        return self.department

    def get_job(self):

        return self.job

    def get_salary(self):

        return self.salary



    def __str__(self):

        return "Name: " +self.name+ "\nID Number "+str(self.number)+"\nDepartment "+self.department+"\nJob "+self.job





    def promote(self):

        bonus = self.salary*0.45

        print(f"Bonus to be offered = {bonus}")



        self.salary *= 1.45

        print(f"Updated Salary = {self.salary}")





wolverine = Employee("Wolverine",1234,"SAP", "Nothing",400000)



print(wolverine)

wolverine.promote()

 
```

```python

'''Q4            '''



class Book:

    def __init__(self,book_name,author_name):

        self.book_name = book_name

        self.author_name = author_name



    def __str__(self):

        return f"{self.book_name} ------- {self.author_name}"





class Library:

    def __init__(self,books_list):

        self.books_list = books_list



    def add_book(self,book):

        self.books_list.append(book)



    def borrow_book(self,book):



        for i in self.books_list:

            if i.book_name == book:

                self.books_list.remove(i)

                print(f"You have rented the book {book}")

                break

        else:

            print(f"Mentioned book {book} is not available in the Library")





    def available_books(self):

        for book in self.books_list:

            print(book)





book1 = Book("Book1","Author1")

book2 = Book("Book2","Author2")

book3 = Book("Book3","Author3")

book4 = Book("Book4","Author4")

book5 = Book("Book5","Author5")



books_list = [book1,book2,book5,book4,book3]



library = Library(books_list)

library.available_books()

library.add_book(Book("Book6", "Author6"))

library.available_books()

library.borrow_book("Book2")

library.available_books()

 
```

```python
 

'''Q5           '''

 



class ProductInventory:



    def __init__(self, products):

        self.products = products



    def add_product(self,product):

        self.products[product] = self.products.get(product,0) + 1



    def update_quantity(self,product,diff_value):

        curr_value = self.products.get(product,0)

        new_value = curr_value + diff_value



        self.products[product] = new_value



    def available_products(self):

        print(self.products)





dict = {

    "Product1": 1,

    "Product2" : 2,

    "Product3" : 3,

    "Product4" : 4,

    "Product5" : 5

}



inventory = ProductInventory(dict)

inventory.available_products()

inventory.add_product("Product6")

inventory.available_products()

inventory.update_quantity("Product3",-2)

inventory.update_quantity("Product6", 2)

inventory.available_products()

 

```

```python

'''Q6              '''



class Shape:



    def __init__(self,length,width,height):

        self.length = length

        self.width = width

        self.height = height



    def area(self):



        total_area = 2 * (self.length + self.width +self.height)



        print(f"Total Surface Area is {total_area}")



    def perimeter(self):



        total_perimeter = 4 * self.length + 4 * self.width + 4 * self.height



        print(f"Total perimeter is {total_perimeter}")





shape = Shape(10,20,30)

shape.area()

shape.perimeter()

 
```

```python

'''Q7               '''





class Student:



    def __init__(self,id,name,grades):

        self.id = id

        self.name = name

        self.grades = grades



    def avg_score(self):

        sum = 0

        count = 0



        for score in self.grades.values():

            sum += score

            count += 1



        print(f"Average Score of {self.name} is {sum/count}")



    def details(self):

        print(f'''

            {self.id} ------- {self.name}

            {self.grades}

        ''')





grades = {

    "Subject 1" : 90,

    "Subject 2" : 80,

    "Subject 3" : 95

}



student = Student(101, "Wolverine", grades)



student.details()

student.avg_score()

 
```

```python
 

'''Q10           '''



class ToDoList:



    def __init__(self,tasks):

        self.tasks = tasks



    def add_task(self,task):

        self.tasks.append(task)



    def mark_completed(self,task_name):

        print(f"{task_name} is completed")

        for task in self.tasks:

            if task[0] == task_name:

                self.tasks.remove(task)

                break

        else:

            print("No such task in ToDoList")



    def display(self):

        print(self.tasks)





tasks = [["task1", 'date1'],

         ["task2", 'date2'],

         ["task3", 'date3'],

         ["task4", 'date4'],

         ["task5", 'date5']]



todo = ToDoList(tasks)

todo.display()

todo.add_task(["newTask", "NewDate"])

todo.display()

todo.mark_completed("task4")

todo.display()

```
